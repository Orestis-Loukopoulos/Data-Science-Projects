# Task A

```python
#import libraries
import pandas as pd
import numpy as np
import pprint
import networkx as nx
from itertools import combinations
```

```python
#load csv
mydata = pd.read_csv('ER-Data.csv', delimiter=';')
mydata
```

```python
#replace NaN with None
mydata = mydata.where(pd.notnull(mydata), None)
```

```python
#lowercase the values for columns that contain characters
mydata.authors = mydata.authors.str.lower()
mydata.venue = mydata.venue.str.lower()
mydata.title = mydata.title.str.lower()
mydata
```

```python
#Removed non-alphanumeric characters from strings within every dataframe column

mydata.authors = mydata.authors.str.replace('[^a-zA-Z0-9]', ' ')
mydata.authors = mydata.authors.str.replace('\s+',' ') #double space to single space
mydata.authors = mydata.authors.str.rstrip() #remove whitespace from the end of the string

mydata.venue = mydata.venue.str.replace('[^a-zA-Z0-9]', ' ')
mydata.venue = mydata.venue.str.replace('\s+',' ') #double space to single space
mydata.venue = mydata.venue.str.rstrip() #remove whitespace from the end of the string

mydata.title = mydata.title.str.replace('[^a-zA-Z0-9]', ' ')
mydata.title = mydata.title.str.replace('\s+',' ') #double space to single space
mydata.title = mydata.title.str.rstrip() #remove whitespace from the end of the string

mydata
```

```python
#split to list of words

mydata.authors = mydata.authors.str.split(" ", expand=False)
mydata.venue = mydata.venue.str.split(" ", expand=False)
mydata.title = mydata.title.str.split(" ", expand=False)

mydata
```

```python
blocks = {} #initialize an empty dictionary

for i in range(len(mydata)): #iterate through rows
    
    for j in range(1, mydata.shape[1]): #iterate through columns
        
        if type(mydata.iloc[i, j]) == list: #columns that contain lists
            for k in list(mydata.iloc[i, j]): #iterate through list
                index = 'Entity_' + str(i)
                if k not in blocks.keys(): #check if the token already exists in the block as key
                    blocks[k] = []
                blocks[k].append(index) #append the entity number in the corresponding token (block index) 
                
        else: #non lists
            m = mydata.iloc[i, j]
            if m == None:
                    continue
            index = 'Entity_' + str(i)
            if m not in blocks.keys(): #check if the token already exists in the block as key
                blocks[m] = []  
            blocks[m].append(index) #append the entity number in the corresponding token (block index)

    
```

```python
#Pretty print the indeces (keys) of the blocks

pprint.pprint(blocks.keys())
```

# Task C

```python
#For each token (key) in the blocks dictionary, find all the possible combinations of Entities and append it to a list
#This list will represent the existing edges between Entities.
#We should mention that 'list edges' is a list of lists which contains tuples. 
#Each tuple is a pair of Entities that appear in the same block.

edges = []

for i in list(blocks.keys())[:10]: #take only the first 10 keys for computational reasons
    a = list(combinations(blocks[i], 2))
    edges.append(a)
```

```python
#Append each tuple into a list.
#So the new list contains all the pairs of Entities.

list_of_edges = [] #initialize an empty list which will contain all the edges
for i in range(len(edges)):
    for j in range(len(edges[i])):
            list_of_edges.append(edges[i][j])
```

```python
new_df = pd.DataFrame(list_of_edges) #list to dataframe
new_df.columns = ['Node_1', 'Node_2']
new_df['Weight'] = np.ones(len(list_of_edges)) #Initializing column Weights with ones.
    
#Calculate the Weights
#If a row has Node_1: Entity_10 and Node_2: Entity_25 and another row has Node_1: Entity_25 and Node_2: Entity_10 (Symmetry)
#then we will count their appearance with weight equal to 2 in one row and not 1 in each row.
#In different words we take into account that a symmetric property exists between columns Node_1 and Node_2.

final_df = (new_df.groupby(new_df[['Node_1', 'Node_2']].apply(lambda x: str(sorted(x)), axis=1))
              .agg({"Node_1": "first", "Node_2": "first", "Weight": "sum"}).reset_index(drop=True))

final_df
```

```python
edge_list = final_df.values.tolist() #convert dataframe to a list that contains lists (list of lists)
edge_list
```

```python
G = nx.DiGraph() #define the graph

for i in range(len(edge_list)):
    G.add_edge(edge_list[i][0], edge_list[i][1], weight = edge_list[i][2]) #add edges on the graph
```

```python
#remove edges with weight < 2

threshold = 2

# filter out all edges above threshold
unwanted_edges = list(filter(lambda e: e[2] < threshold, (e for e in G.edges.data('weight'))))
ids = list(e[:2] for e in unwanted_edges) #isolate the ids of edges with weight < 2

# remove filtered edges from graph G
G.remove_edges_from(ids)
```

```python
#Calculate the number of comparisons for the subset of 10 Blocks.
#We will count the rows of the dataframe which have weight greater than or equal to 2.

len(final_df.loc[final_df.Weight>=2])
```

```python
#define Jaccard Similarity function

def jaccard(id_1, id_2):
    title1 = mydata.title.loc[id_1]
    title2 = mydata.title.loc[id_2]
    intersection = len(list(set(title1).intersection(title2)))
    union = (len(title1) + len(title2)) - intersection
    jac = float(intersection) / union
    return jac
```

```python
# Entities to calculate their Jaccard Similarity

#Give index of each Entity
Entity1 = 1 
Entity3 = 3
```

```python
#Calculate the Jaccard Similarity of the above two Entities

print('The Jaccard similarity of the above Entities is equal to {}'
      .format(jaccard(Entity1, Entity3)))
```
